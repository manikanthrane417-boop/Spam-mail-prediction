```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pickle

df=pd.read_csv(r"/Users/vaibhav/Desktop/AI_project/spam_mail_prediction/email.csv")
print(df.shape)
print(df.head())
print(df['Category'].value_counts())
```

```python
labelln=LabelEncoder()
label=labelln.fit_transform(df['Category'])
df['target']=label
x=df.Message
y=df.target
print(x)
print(y)
```

```python
x_train,x_text,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
print(x.shape)
print(x_train.shape)
print(x_text.shape)
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english')
x_train_feature=feature_extraction.fit_transform(x_train)
x_test_feature=feature_extraction.transform(x_text)
print(x_train)
print(x_train_feature.toarray())
```

```python
model=MultinomialNB()
model.fit(x_train_feature,y_train)
#prediction on training data
prediction_on_training_data=model.predict(x_train_feature)
accuracy_score_on_trian=accuracy_score(y_train,prediction_on_training_data)
print(f"Accuracy on traindata: {accuracy_score_on_trian}")
#prediction on testing data
prediction_on_testing_data=model.predict(x_test_feature)
accuracy_score_on_test=accuracy_score(y_test,prediction_on_testing_data)
print(f"Accuracy on testdata: {accuracy_score_on_test}")
```

```python
#save files
with open("model.pkl",'wb') as f:
    pickle.dump(model,f)

with open("vectorer.pkl","wb") as f:
    pickle.dump(feature_extraction,f)

print("saved succesfully")
```
